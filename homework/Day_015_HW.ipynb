## np.random.random_sample

### (b - a) * random_sample() + a
 To sample Unif[a, b), b > a multiply the output of random_sample by (b-a) and add a


```python
#請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
"""
Your Code Here
"""


matrix = 2*np.random.random_sample((10,10))-1

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)

plt.show()
```

## np.random.randn()


For random samples from N(\mu, \sigma^2), use:
#### sigma * np.random.randn(...) + mu


```python
#請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = 1 * np.random.randn(1000,3)-1

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
# 繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randn(1000,3)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
# 繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
